```python
from ultralytics import YOLO
```

# Data Collection & Preprocessing

```python
Marine_img=r"D:/Data centr/IMG_data/Marine_Ship/data.yaml",
```

# Pre trained Model Selection

```python
model= YOLO("yolov8n.pt")
```

# Model training on custom data

```python
trained_model = model.train(
    data=Marine_img,
    epochs=40,
    imgsz=640,
    batch=8,
    workers=4,
    
    # Optimization
    optimizer="AdamW",
    lr0=0.003, # lower LR for fine-tuning
    weight_decay=0.0005,
    
    # Small Object Focus
    mosaic=1.0,
    close_mosaic=10,
    mixup=0.1, # mild mixup helps robustness
    copy_paste=0.0,
    
    # Augmentation
    hsv_h=0.015,
    hsv_s=0.4,
    hsv_v=0.4,
    fliplr=0.5,
    flipud=0.0, # ships rarely upside-down in SAR
    
    # Stability
    patience=20,
    cos_lr=True,
    
    # Validation
    val=True
)
```

# Model Evaluation
